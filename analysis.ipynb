## Import Libraries Required

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Analysis of Value Iteration States and Utilities

##### (ε = 0.1)
ε is the maximum error allowed in the utility of any state. It is used to calculate `threshold` for termination of value iteration algorithm. 

```python
VI_FILE_PATH = "docs\\analysis\\value_iteration_assgn1.csv"
vi_data = pd.read_csv(VI_FILE_PATH)
print(vi_data.shape)
vi_data.head()
```

_Note: One may think that there should be 36 columns (one associated with each state of 6x6 grid world). However, since 5 of the states in the grid world are Walls, there is no utility associated with them. Hence, they have not been included in the csv file_

```python
X = np.linspace(1, vi_data.shape[0], vi_data.shape[0])

fig, ax = plt.subplots(figsize=(15, 10))

for state in vi_data:
    ax.plot(X, vi_data[state], label=str(state))
ax.set_title('Value Iteration - Utility vs. Iterations (ε = 0.1)', size=22)
ax.set_xlabel('Iteration #', size=18)
ax.set_ylabel('Utility of State', size=18)
ax.legend(loc="lower right", ncols=3)

fig.show()
fig.savefig(VI_FILE_PATH.replace('csv', 'jpg'))
```
